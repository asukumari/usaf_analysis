## Comparison of algorithms
Let's use all three algorithms and see if we can notice a difference...

```python
fig, axes = plt.subplots(1,3, figsize=(16,6))
f2, ax2 = plt.subplots(1,1, figsize=(16,6))
for i, demosaic in enumerate((
    demosaicing_CFA_Bayer_bilinear,
    demosaicing_CFA_Bayer_Malvar2004,
    demosaicing_CFA_Bayer_Menon2007,)):
    debayered = demosaic(usaf_bayer/(2**10 - 1))
    axes[i].imshow(debayered[1100:1300,1750:1950,...]#/np.max(np.max(debayered,axis=0),axis=0))
    ax2.plot(debayered[1200,1650:2050,:])
    
```

## Simple linear implementation
Just to make sure I know what's going on, here is a simple bilinear algorithm from me...

```python
bayer = np.zeros((20,20,3), dtype=np.uint16)
(ry, rx), (gy, gx), (Gy, Gx), (by, bx) = ((0, 0), (1, 0), (0, 1), (1, 1))
bayer[ry::2, rx::2, 0] = 1 # Red
bayer[gy::2, gx::2, 1] = 1 # Green
bayer[Gy::2, Gx::2, 1] = 1 # Green
bayer[by::2, bx::2, 2] = 1 # Blue
bayer *= 2**10 -1
plt.imshow((bayer/4).astype(np.uint8))
```

```python
convolve = scipy.ndimage.filters.convolve
arr = bayer.copy()
out = np.zeros_like(arr) # if we try to do it in-place, we make a nasty mess...
for i in range(3):
    weights = np.array([[1,1,1]],dtype=np.uint16)
    convolve(arr[..., i], weights, output=out[..., i], mode="constant")
arr = out // 4
plt.imshow((arr/4).astype(np.uint8))
```
